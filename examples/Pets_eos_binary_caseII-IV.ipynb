```python
petsII  = Pets(PetsParameters.from_lists(epsilon_k=[epsilon_k/KELVIN]*2,                     sigma=[sigma/ANGSTROM]*2, k_ij=np.array([[0.0, 0.15], [0.15, 0.0]])))
petsIII = Pets(PetsParameters.from_lists(epsilon_k=[epsilon_k/KELVIN]*2,                     sigma=[sigma/ANGSTROM]*2, k_ij=np.array([[0.0, -0.25], [-0.25, 0.0]])))
petsIV  = Pets(PetsParameters.from_lists(epsilon_k=[0.6*epsilon_k/KELVIN, epsilon_k/KELVIN], sigma=[sigma/ANGSTROM]*2))
```
